Data Imputation

```python
fig, axes = plt.subplots(nrows = 4, ncols = 2, figsize = (10, 10))
axes = axes.flatten()

for i, feature in enumerate(df.columns):
  sns.histplot(data = df[feature], kde = True, ax = axes[i])
  axes[0].set_title(f"{feature}")

  if(i == 7):
    break

plt.tight_layout()
plt.show()

plt.savefig("featur_distribution.png")
```

```python
df['Pregnancies']               = df['Pregnancies'].replace(0, df['Pregnancies'].median())
df['SkinThickness']             = df['SkinThickness'].replace(0, df['SkinThickness'].median())
df['Insulin']                   = df['Insulin'].replace(0, df['Insulin'].median())
df['DiabetesPedigreeFunction']  = df['DiabetesPedigreeFunction'].replace(0, df['DiabetesPedigreeFunction'].median())
df['Age']                       = df['Age'].replace(0, df['Age'].median())
df['Glucose']                   = df['Glucose'].replace(0, df['Glucose'].mean())
df['BloodPressure']             = df['BloodPressure'].replace(0, df['BloodPressure'].mean())
df['BMI']                       = df['BMI'].replace(0, df['BMI'].mean())
```

## Outlier Detection

```python
X = df.drop(columns='Outcome', axis = 1)
Y = df['Outcome']
```

```python
# Box plot for all features initially

fig, ax = plt.subplots(figsize = (15,15))
sns.boxplot(X, ax = ax)
plt.show()
plt.savefig("boxplot.jpg")
```

```python
# Handling Outliers

for feature in X.columns:
    Q1 = X[feature].quantile(0.25)
    Q3 = X[feature].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Replace outliers with NaN
    X[feature] = X[feature].mask((X[feature] < lower_bound) | (X[feature] > upper_bound))

# Impute NaNs with column median
X = X.fillna(X.median())
```

```python
# Box plot for all features after outlier removal

fig, ax = plt.subplots(figsize = (15,10))
sns.boxplot(X, ax = ax)
plt.show()
plt.savefig("boxplot_after.jpg")
```

#### Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_norm = scaler.fit_transform(X)

X_norm = pd.DataFrame(X_norm, columns= X.columns)
```

```python
# Box plot for all features after standardization

fig, ax = plt.subplots(figsize = (15,10))
sns.boxplot(X_norm, ax = ax)
plt.show()
plt.savefig("boxplot_norm.jpg")
```
